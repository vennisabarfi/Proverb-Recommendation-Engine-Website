```python
from bs4 import BeautifulSoup
import requests as re
```

```python
url = "http://swahiliproverbs.afrst.illinois.edu/proverbs.htm"
page = re.get(url)
soup = BeautifulSoup(page.content, "html.parser")

print(page.text)
```

```python
results= soup.find(id ="centerContent")
```

```python
print(results.prettify())
```

```python
proverbs = results.find_all("span", class_ = "style11")
print(proverbs)
```

```python
# soup2 = BeautifulSoup(proverb_meanings, "html.parser")
# proverb_meanings = results.find_all("span", class_ = "style9")
```

```python
line_output = "".join(tag.text for tag in proverbs)
lo = line_output.replace("Meaning:", " ")
l1 = lo.replace("Application:", " ")
print(l1)
```

```python
import json
```

```python
# write to text file
with open("proverbs_1", "w") as f:
    f.write(l1)
```

```python
# write to json file
# write to a json file
with open("proverbs_1_json", "w") as json_file:
    json.dump(l1, json_file, indent=4)
```

```python
# write to jsonl file
with open("proverbs_1_jsonl", "w") as jsonl_file:
    for proverb in l1:
        jsonl_file.write(json.dumps(proverb) + "\n")
```
